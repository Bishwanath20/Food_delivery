```python
import pandas as pd
import sqlite3
 
```

```python
pd.set_option('display.max_columns', None)
print("Libraries imported successfully!")
```

```python

df_orders = pd.read_csv('orders.csv')
df_users = pd.read_json('users.json')
```

```python
print(f"Orders loaded: {df_orders.shape[0]} rows")
print(f"Users loaded: {df_users.shape[0]} rows")
```

```python

conn = sqlite3.connect(':memory:')
with open('restaurants.sql', 'r') as file:
    sql_script = file.read()
conn.executescript(sql_script)
df_restaurants = pd.read_sql_query("SELECT * FROM restaurants", conn)
```

```python
print(f"Restaurants loaded: {df_restaurants.shape[0]} rows")
```

```python
df_merged = pd.merge(df_orders, df_users, on='user_id', how='left')
final_df = pd.merge(df_merged, df_restaurants, on='restaurant_id', how='left')

```

```python
print("Merge complete! First 5 rows:")
display(final_df.head())
```

```python
final_df.to_csv('final_food_delivery_dataset.csv', index=False)
print("\nSuccess! 'final_food_delivery_dataset.csv' has been saved in your folder.")
```
